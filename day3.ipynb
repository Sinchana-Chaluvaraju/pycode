```python
import qrcode
import image
```

```python
qr = qrcode.QRCode(
    version = 5,
    box_size = 10,
    border = 5
)

data = "https://www.youtube.com/watch?v=eh0RH-riAOU"
```

```python
qr.add_data(data)
```

```python
qr.make_image(fit = True)
```

```python
img = qr.make_image(fill = "black", back_color = "white")
```

```python
img.save("test.png")
```
